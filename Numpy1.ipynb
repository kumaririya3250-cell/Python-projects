# pseudo seed numbers
### np.random.seed()
#### np.random.seed() is used to control the randomness so that you get the same random numbers every time you run the code.

### Normally, random numbers change every run.
### But when you set a seed, NumPy starts the random generator from the same starting point, so the output becomes repeatable.

```python
## np.random.seed(999999)
random_array4 = np.random.random((10,5))
random_array4
```

```python
np.random.seed(999999)
random_array_5 = np.random.randint(30, size = (9,11))
random_array_5
```

```python
##seed() is used for:
#1️⃣ Reproducible results
#Data science experiments
#Machine learning models
#2️⃣ Debugging code
#So the random numbers don't change every run.
#3️⃣ Sharing research/code
#Other people can get the exact same result.
```

```python
np.random.seed(7)
random_array_5 = np.random.randint(30, size = (9,11))
random_array_5
#Same seed → same random numbers
#Different seed → different random numbers
```

# You need to use that transpose (.T) you were experimenting with to flip one of the matrices so the dimensions align.
## If you want the result to be a (3X3) matrix, transpose the first matrix:
## (3,5) dot (5,3) -> result is (3, 3)

```python
np.dot(matrix_2.T,matrix_1)
```

# Alternatively, if you want a (5X5) matrix, transpose the second
## (5, 3) dot (3, 5) -> result is (5, 5)

```python
np.dot(matrix_2, matrix_1.T)
```

```python
matrix_3 = matrix_1.T
```

```python
matrix_1, matrix_3
```

```python
np.dot(matrix_2,matrix_3)
```

```python
print(np.dot(matrix_2,matrix_3))
```

## Numpy.argsort : returns the indices that would sort an array, Perform an indirect sort along the given axis using the algorithm specified by the kind keyword
## np.argsort(a,axis=-l,kind=none,order=none,*,stable=none)
## Return the indices that would sort an array

### Perform an indirect sort along the given axis using the algorithm specified by the kind keyboard. It returns an array of indices of the same shape as a that index data along the given axis in sorted order
