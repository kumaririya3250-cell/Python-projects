```python
print("Calculator")
a = int(input("Enter the number"))
b = int(input("Enter the number"))
operator = input("Enter the operator(+,-,*,/,%,//,**):")

if operator == "+":
    print("Result:", a+b)
elif operator == "-":
    print("Result:", a-b) 
elif operator == "*":
    print("Result:", a*b)
elif operator == "/":
    print("Result:", a/b)
elif operator == "%":
    print("Result:", a%b)
elif operator == "//":
    print("Result:", a//b)
elif operator == "**":
    print("Result:", a**b)
else:
    print("Invalade Operator")
```
